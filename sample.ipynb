```python
import requests

api_key = '45b371ace66f4f11b97f83cce88fc33d'

url = 'https://newsapi.org/v2/everything'  # From Newsapi.org

params = {
    'q': "Foreign trade",      # Your search query
    'language': 'en',
    'sortBy': 'publishedAt',
    'apiKey': api_key
}

response = requests.get(url, params=params)
news_data = response.json()
print(news_data.get('articles'))
```

# Extracting URL's

```python
articles = news_data['articles']

for article in articles:
    print(f"Title: {article['title']}")
    print(f"Source: {article['source']['name']}")
    print(f"Published At: {article['publishedAt']}")
    print(f"URL: {article['url']}")
    print("\n" + "="*50 + "\n")
```

# Extracting Title

```python
import requests
import trafilatura

for article in articles:
    article_url = article["url"]

    html = requests.get(article_url, headers={'User-Agent': 'Mozilla/5.0'}).text
    text = trafilatura.extract(html)
    metadata = trafilatura.extract_metadata(html, article_url)

    print("URL:", article_url)
    print("Title:", metadata.title)
    print("Date:", metadata.date)
    print("Content Preview:", text[:800])
```
